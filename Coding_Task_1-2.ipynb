```python
import pandas as pd


df = pd.read_excel('Dataset.xlsx')


data_bulanan = []  


temp_data = {
    'Tanggal': [],
    'Curah Hujan': [],
    'Penjualan A': [],
    'Penjualan B': []
}

for index, row in df.iterrows():
    temp_data['Tanggal'].append(row['Tanggal'])
    temp_data['Curah Hujan'].append(row['Curah Hujan (mm)'])
    temp_data['Penjualan A'].append(row['Penjualan A (pcs)'])
    temp_data['Penjualan B'].append(row['Penjualan B (pcs)'])
    
    
    if (index + 1 < len(df) and df['Tanggal'][index + 1] == 1) or (index + 1 == len(df)):
        data_bulanan.append(temp_data.copy())  
        temp_data = {  
            'Tanggal': [],
            'Curah Hujan': [],
            'Penjualan A': [],
            'Penjualan B': []
        }
```

```python
import statistics as st
import numpy as np

All_A = [] 
All_B = []
cH = []

print("Rata-rata per bulan")
for i, data in enumerate(data_bulanan):
    All_A.extend(data['Penjualan A'])
    All_B.extend(data['Penjualan B'])
    cH.extend(data['Curah Hujan'])
    
    
    meanA = st.mean(data['Penjualan A'])
    meanB = st.mean(data['Penjualan B'])

    
    print(f"Rata-rata penjualan A pada bulan {i+1} = {meanA}")
    print(f"Rata-rata penjualan B pada bulan {i+1} = {meanB}\n")



print(f"Rata-rata Penjualan Barang A = {st.mean(All_A)}")
print(f"Rata-rata Penjualan Barang B = {st.mean(All_B)}\n")

print(f"Modus Barang A = {st.multimode(All_A)}")
print("Modus per Bulan:")
for i, data in enumerate(data_bulanan):
    modusA = st.multimode(data['Penjualan A'])
    print(f"Modus Penjualan barang A pada Bulan {i+1} = {modusA} ")

print(f"Modus Barang B = {st.multimode(All_B)}")
print("Modus per Bulan: ")
for i, data in enumerate(data_bulanan):
    modusB = st.multimode(data['Penjualan B'])
    print(f"Modus Penjualan barang B pada Bulan {i+1} = {modusB} ")

print(cH)
```

```python
import statistics as st
import numpy as np
from scipy.stats import skew, kurtosis



# Median per bulan
for i, data in enumerate(data_bulanan):
    print(f"Median Barang A Bulan ke-{i + 1} = {st.median(data['Penjualan A'])}")
    print(f"Median Barang B Bulan ke-{i + 1} = {st.median(data['Penjualan B'])}\n")

# Median keseluruhan
print(f"Median Barang A (Keseluruhan) = {st.median(All_A)}")
print(f"Median Barang B (Keseluruhan) = {st.median(All_B)}\n")

# Range (selisih antara nilai maksimum dan minimum)
print(f"Range Data A = {max(All_A) - min(All_A)}")
print(f"Range Data B = {max(All_B) - min(All_B)}\n")

# Quartil dan Interquartil
print("Quartil dan Interquartil Data A: ")
print(f"Quartil 1 = {np.percentile(All_A, 25)}")
print(f"Quartil 2 (Median) = {np.percentile(All_A, 50)}")
print(f"Quartil 3 = {np.percentile(All_A, 75)}")
print(f"Interquartil Range = {np.percentile(All_A, 75) - np.percentile(All_A, 25)}\n")

print("Quartil dan Interquartil Data B: ")
print(f"Quartil 1 = {np.percentile(All_B, 25)}")
print(f"Quartil 2 (Median) = {np.percentile(All_B, 50)}")
print(f"Quartil 3 = {np.percentile(All_B, 75)}")
print(f"Interquartil Range = {np.percentile(All_B, 75) - np.percentile(All_B, 25)}\n")

# Varians
print(f"Varians Produk A = {st.variance(All_A)}")
print(f"Varians Produk B = {st.variance(All_B)}\n")

# Standar Deviasi
print(f"Standar Deviasi Produk A = {np.std(All_A, ddof=1)}")
print(f"Standar Deviasi Produk B = {np.std(All_B, ddof=1)}\n")

# Skewness (mengukur kemiringan distribusi)
print(f"\nSkewness Produk A = {skew(All_A)}")
print(f"Skewness Produk B = {skew(All_B)}")

#Kurtosis
print(f"\nKurtosis Produk A = {kurtosis(All_A)}")
print(f"Kurtosis Produk B = {kurtosis(All_B)}")
```

```python
import matplotlib.pyplot as plt

#Histogram Curah Hujan
cH_low = min(cH)
cH_max = max(cH)

interval_cH = cH_max - cH_low

c_cH = interval_cH // 8

bins = np.arange(cH_low, cH_max + c_cH, c_cH)

plt.hist(cH, bins=bins, edgecolor='black')

plt.title('Histogram Distribusi Frekuensi Curah Hujan')
plt.xlabel('Curah Hujan')
plt.ylabel('Frekuensi')

plt.show()

#Histogram Penjualan A
A_low = min(All_A)
A_max = max(All_A)

interval_A = A_max - A_low

c_A = interval_A // 8

bins = np.arange(A_low, A_max + c_A, c_A)

plt.hist(All_A, bins=bins, edgecolor='black')

plt.title('Histogram Distribusi Frekuensi Penjualan A')
plt.xlabel('Penjualan A (Pcs)')
plt.ylabel('Frekuensi')

plt.show()

#Histogram Penjualan B

B_low = min(All_B)
B_max = max(All_B)

interval_A = B_max - B_low

c_B = interval_A // 8

bins = np.arange(B_low, B_max + c_B, c_B)

plt.hist(All_B, bins=bins, edgecolor='black')

plt.title('Histogram Distribusi Frekuensi Penjualan B')
plt.xlabel('Penjualan B (Pcs)')
plt.ylabel('Frekuensi')

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


data_hujan_AB = [{'Curah Hujan': data['Curah Hujan'], 'Penjualan A': data['Penjualan A'], 'Penjualan B': data['Penjualan B']} for data in data_bulanan]


curah_hujan = [entry['Curah Hujan'] for entry in data_hujan_AB]
penjualan_A = [entry['Penjualan A'] for entry in data_hujan_AB]
penjualan_B = [entry['Penjualan B'] for entry in data_hujan_AB]

curah_hujan_flat = [nilai for sublist in curah_hujan for nilai in sublist]
penjualan_A_flat = [nilai for sublist in penjualan_A for nilai in sublist]
penjualan_B_flat = [nilai for sublist in penjualan_B for nilai in sublist]

bins = np.linspace(min(curah_hujan_flat), max(curah_hujan_flat), num=9)  

data = pd.DataFrame({'Curah Hujan': curah_hujan_flat, 'Penjualan A': penjualan_A_flat, 'Penjualan B': penjualan_B_flat})
data['Kategori Curah Hujan'] = pd.cut(data['Curah Hujan'], bins=bins, right=False)

penjualan_A_grouped = data.groupby('Kategori Curah Hujan')['Penjualan A'].mean()
penjualan_B_grouped = data.groupby('Kategori Curah Hujan')['Penjualan B'].mean()

midpoints = [(interval.left + interval.right) / 2 for interval in penjualan_A_grouped.index]

plt.figure(figsize=(10, 6))
plt.plot(midpoints, penjualan_A_grouped, marker='o', linestyle='-', color='orange', label='Penjualan A')
plt.plot(midpoints, penjualan_B_grouped, marker='o', linestyle='-', color='blue', label='Penjualan B')

plt.title('Poligon Frekuensi Penjualan Berdasarkan Curah Hujan')
plt.xlabel('Curah Hujan (mm)')
plt.ylabel('Rata-rata Penjualan')
plt.grid(True)
plt.legend()

plt.show()
```

<b>Interpretasi Outlier:</b>
Outlier adalah data yang sangat berbeda atau jauh dari mayoritas data lainnya. Dalam dataset yang diberikan, boxplot dari distribusi curah hujan serta penjualan produk A dan B menunjukkan adanya outlier, terutama pada data curah hujan yang mungkin dipengaruhi oleh kondisi cuaca ekstrem. Pada data penjualan produk A dan B, terlihat beberapa titik yang ekstrem, namun outlier-outlier ini tidak memberikan informasi kritis yang signifikan dan bisa diabaikan dalam konteks analisis umum.
